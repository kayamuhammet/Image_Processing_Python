```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt

img = cv.imread('C:\Image\wiki.jpg',0)
cv.imshow("Image", img)

#Computes the histogram of the image and its values.
hist,bins = np.histogram(img.flatten(),256,[0,256])

cdf = hist.cumsum() #Calculates the cumulative distribution function.

#Computes the normalized cumulative distribution function.
cdf_normalized = cdf * float(hist.max()) / cdf.max()

#Plots the normalized cumulative distribution function in blue.
plt.plot(cdf_normalized, color = 'b')

#Plots the histogram of the image in red.
plt.hist(img.flatten(),256,[0,256], color = 'r')

plt.xlim([0,256]) #Sets the limits for the x-axis.

#Adds a legend specifying what the lines in the plot represent.
plt.legend(('cdf','histogram'), loc = 'upper left')

plt.show() # Displays the created plots.

cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# Reads the image 'wiki.jpg' in grayscale mode from the specified file path.
img = cv.imread('C:\Image\wiki.jpg',0)

# Performs histogram equalization on the input grayscale image.
equ = cv.equalizeHist(img)

# Stacks the original image and the equalized image side by side and converts the result to the uint8 data type.
res = np.hstack((img,equ)).astype(np.uint8) 

cv.imwrite('res.png',res)
```
